```python
nufus=pd.read_csv("dunyam.csv")
nufus.dropna(how="any",inplace=True)
nufus.head()
```

```python
sns.set(style="dark",palette="muted",font_scale=2)
sns.distplot(nufus["Doğum oranı"],bins=20,kde=False,color="y")
plt.tight_layout()
```

```python
sns.set(style="darkgrid",palette="muted",font_scale=1.5)
sns.distplot(nufus["Yıllık değişim oranı"],hist=False,rug=True,color="r")
plt.tight_layout()
```

```python
sns.set(style="white",palette="Blues",font_scale=1.5)
sns.distplot(nufus["Ortalama yaş"],hist=False,color="g",kde_kws={"shade":True})
plt.tight_layout()
```
