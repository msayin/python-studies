```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
x= ["1.USA","2.CHINA","3.JAPAN","4.GERMANY","5.ENGLAND","6.FRANCE"]
y= [20410,14090,5170,4210,2940,2930] # GSMH Gelirler
z= [63039,9957,40649,51158,44162,44915] # Kisi basina dusen milli gelir
```

```python
# ulkelerin GSMH gore olusturuldu
plt.plot(x,y)
```

```python
# Burada kisi basi gelire gore olusturuldu
plt.plot(x,z)
```

```python
# Rekler atamak, istediginiz rengin kodunu yaziyorsunuz
plt.plot(x,z,"r")
```

```python
# Burada x,y label ve title atamis olduk
plt.plot(x,y,"g")
plt.xlabel("Countries")
plt.ylabel("GDP")
plt.title("Countries GDP")
```

```python
# ic ice grafik olusturmak
plt.subplot(1,2,1)
plt.plot(x,y,"r--") # burada kesik cizgi ekledik
plt.subplot(1,2,2)
plt.plot(x,z,"y*-") # Bura da da * - cizgi eklemis olduk
plt.tight_layout() # Burada iki frafik arasinda mesafa koymus olduk
```

```python
f=plt.figure ()
axes=f.add_axes([0.1,0.1,0.9,0.9]) # Burada frafigin cercevesi icin boyutlarini vererek bir tablo olusturduk
axes.plot(x,y ,"r")
axes.set_xlabel("Countries")
axes.set_ylabel("GDP")
axes.set_title("Countries GDP")
```

```python
# ic ice grafik olusturma
f=plt.figure ()
axes1=f.add_axes([0.1,0.1,0.9,0.9]) # Burada frafigin cercevesi icin boyutlarini vererek bir tablo olusturduk
axes2=f.add_axes([0.5,0.5,0.4,0.4])
axes1.plot(x,y ,"r")
axes2.plot(x,z ,"y")
plt.tight_layout()
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2) # bu durumda satir ve sutun ekliyoruz
# bu birinci grafik
axes[0].plot(x,y,"b")
axes[0].set_title("GDP")

# bu ikinci grafik
axes[1].plot(x,z,"y")
axes[1].set_title("Kisi basi GDP")
plt.tight_layout()# iki grafik arasinda uzaklik olusturuyor      
```

```python
f=plt.figure ()
axes=f.add_axes([0.1,0.1,0.9,0.9]) # Burada frafigin cercevesi icin boyutlarini vererek bir tablo olusturduk
axes.plot(x,y,color="red", 
    linestyle="-.", # burada cizgi olsun dedik
    linewidth=5, # burada cizginin kalinligini belirledik
    marker="o", # burada daireler ekledik
    markersize=20, # burada dairelere boyutunu belirledik
    markerfacecolor="yellow", #burada dairelere rengini belirledik
    markeredgewidth=4, # burada dairelere kenar kalinligini belirledik
    markeredgecolor="blue", # burada dairelere kenar rengini belirledik
    alpha=0.6) # burada grafigin saydamligini belirledik
axes.set_xlabel("Countries")
axes.set_ylabel("GDP")
axes.set_title("Countries GDP")
```

```python
f=plt.figure ()
axes=f.add_axes([0.1,0.1,0.9,0.9]) # Burada frafigin cercevesi icin boyutlarini vererek bir tablo olusturduk
axes.plot(x,y,color="red")
axes.set_xlabel("Countries")
axes.set_ylabel("GDP")
axes.set_title("Countries GDP")
axes.set_xlim([2,6]) # burada kactane ulkeyi alacaksan onu belirtiyorsun, yani zumlama yapiyorsun
axes.set_ylim([2000,7500])
```

```python
# Grafigin Boyutu degistirme
f=plt.figure(figsize=(10,6)) # burada grafigin boyutunu ayarliyorsunuz 10 sag sol genisletiriyor, 6 asagi yukari genisletiyor
axes=f.add_axes([0,0,1,1]) # Burada frafigin cercevesi icin boyutlarini vererek bir tablo olusturduk
axes.plot(x,y)
axes.set_xlabel("Countries")
axes.set_ylabel("GDP")
axes.set_title("Countries GDP")
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(13,6)) #burada grafigin boyutunu ayarliyorsunuz 10 sag sol genisletiriyor, 6 asagi yukari genisletiyor 
# bu birinci grafik
axes[0].plot(x,y,"b")
axes[0].set_title("GDP")

# bu ikinci grafik
axes[1].plot(x,z,"y")
axes[1].set_title("Kisi basi GDP")
plt.tight_layout()# iki grafik arasinda uzaklik olusturuyor      
```

```python
# olusturdugumuz taplolari resim olarak kaydedebiliriz
fig.savefig("resim.png")
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'TURKEY', 'USA', 'CHINA', 'GERMANY'
sizes = [15, 35, 35, 15]
explode = (0, 0.1, 0, 0)  # burdaki rakamlarla oynayarak isdedigimiz dili ayirabiliyoruz. ornekte oldugu gibi 
                        # Burda USA ayrildi. Bunu index sirasina gore yapiyoruz
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'TURKEY', 'USA', 'CHINA', 'GERMANY'
sizes = [15, 35, 35, 15]
explode = (0.3, 0.0, 0, 0)  # burdaki rakamlarla oynayarak isdedigimiz dili ayirabiliyoruz. ornekte oldugu gibi 
                        # Burda Turkey dilim olarak ayrildi
                        # hepsini "0" yaparsak tum dilimler birlesik oluyor
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
