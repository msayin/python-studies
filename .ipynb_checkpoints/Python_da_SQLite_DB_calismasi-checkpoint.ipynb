```python
import sqlite3
connection = sqlite3.connect("chinook.db")
cursor= connection.execute("select* from customers")
for row in cursor:
    print("First Name = " +row[1])
```

```python
cursor= connection.execute("select FirstName,LastName from customers")
for row in cursor:
    print("First Name = " +row[0])
    print("Last Name = " +row[1])
    print("*************")

```

```python
# burada 'Prague' string olarak gostermemiz ve tek 'tirnak' icine almamiz gerekiyor.
cursor= connection.execute("""select FirstName,LastName 
                            from customers 
                            where city='Prague'""")
#python da 3 tirnak uygulamasi var """ bu sekilde string alt alta
#yazabiliyorsunuz. Ornekte goruldugu gibi
for row in cursor:
    print("First Name = " +row[0])
    print("Last Name = " +row[1])
    print("*************")

```

```python
# "or" fonksiyonunun kullani mi
cursor= connection.execute("""select FirstName,LastName 
                            from customers where 
                            city='Prague' or city='Berlin'""")
for row in cursor:
    print("First Name = " +row[0])
    print("Last Name = " +row[1])
    print("*************")
```

```python
# "orderby" fonksiyonunun kullani mi
#"orderby" ile listeleri A dan Z veya buyukten kucuge dogru siralaya biliyoruz

cursor= connection.execute("""select FirstName,LastName 
                            from customers
                            order by FirstName""")
for row in cursor:
    print("First Name = " +row[0])
    print("Last Name = " +row[1])
    print("*************")
```

```python
# "orderby" fonksiyonunu sondan baslayarak siralamak icin 
# "desc" fonksiyonunu kullaniyoruz.

cursor= connection.execute("""select FirstName,LastName 
                            from customers
                            order by FirstName desc""")
for row in cursor:
    print("First Name = " +row[0])
    print("Last Name = " +row[1])
    print("*************")
```

```python
#"Group by" fonksiyonun kullanimi

cursor= connection.execute("""select city, count (*) 
                            from customers group by city""")
for row in cursor:
    print("City = " +row[0])
    print("Count = " +str(row[1]))
    print("*************")
```

```python
#"Group by" fonksiyonun "order by" ile birlikte kullanimi

cursor= connection.execute("""select city, count (*) 
                            from customers group by city order by count(*)desc""")
for row in cursor:
    print("City = " +row[0])
    print("Count = " +str(row[1]))
    print("*************")
```

```python
#"Group by" fonksiyonun nun "having" order by" ile birlikte kullanimi
# bu sorguda sunu soylemis oluyoruz. Sehre gore groupla sayisi 1 den buyuk olanlari ver.
cursor= connection.execute("""select city, count (*) 
                            from customers group by city having count(*)>1 
                            order by count (*) desc""")
for row in cursor:
    print("City = " +row[0])
    print("Count = " +str(row[1]))
    print("*************")
```

```python
#"Like" fonksiyonun kullanimi gibi 
# bu sorguda icinde "%a%" gecen isimler demek.
cursor= connection.execute("""select FirstName,LastName 
                            from customers
                            where FirstName like '%a%'""")
for row in cursor:
    print("FirstName = " +row[0])
    print("LastName = " +row[1])
    print("*************")
```

```python
#"Like" fonksiyonun kullanimi gibi 
# bu sorguda "a%" ismi "a" ile baslayanlar demek.
cursor= connection.execute("""select FirstName,LastName 
                            from customers
                            where FirstName like 'a%'""")
for row in cursor:
    print("FirstName = " +row[0])
    print("LastName = " +row[1])
    print("*************")
```

```python
#"Like" fonksiyonun kullanimi gibi 
# bu sorguda "%a" ismi "a" ile bitenler demek.
cursor= connection.execute("""select FirstName,LastName 
                            from customers
                            where FirstName like '%a'""")
for row in cursor:
    print("FirstName = " +row[0])
    print("LastName = " +row[1])
    print("*************")
```

```python
#"Like" fonksiyonun kullanimi gibi 
# bu sorguda "%ja" ismi "a" ile bitenler demek.
#yani birden cok degerde verilebilir
cursor= connection.execute("""select FirstName,LastName 
                            from customers
                            where FirstName like '%ja'""")
for row in cursor:
    print("FirstName = " +row[0])
    print("LastName = " +row[1])
    print("*************")
```

```python
#"insert" listeye yeni bir bilgi eklemek icin kullaniyoruz

def insertCustomer():
    connection = sqlite3.connect("chinook.db")
    connection.execute("""insert into customers (FirstName,LastName,city,email)
                        values('Fatih','Sayin','Samsun','fmsayin@hotmail.com')""")
    connection.commit()
insertCustomer()
```

```python
#burada da goruldugu gibi "fatih" ismi bizim db listesine ekelenmis
cursor= connection.execute("""select FirstName,LastName 
                            from customers
                            where FirstName like 'fatih'""")
for row in cursor:
    print("FirstName = " +row[0])
    print("LastName = " +row[1])
    print("*************")
```

```python
#"update" Guncelleme yapmak icin kullaniyoruz
#"insert" listeye yeni bir bilgi eklemek icin kullaniyoruz

def updateCustomer():
    connection = sqlite3.connect("chinook.db")
    connection.execute("""update customers set city= 'istanbul'
                        where city= 'Samsun'""")
    connection.commit()
    
updateCustomer()
```

```python
#burada da goruldugu gibi "istanbul" ismi "samsun" yerine listede guncellenmis
cursor= connection.execute("""select city
                            from customers
                            where city like 'istanbul'""")
for row in cursor:
    print("city = " +row[0])
#     print("city = " +row[1])
    print("*************")
```

```python
#"delete" fonksiyonunun kullanimi
# db. gidip baktigimizda var olan "63" nolu "id" nin silindigini goruruz
def deleteCustomer():
    connection = sqlite3.connect("chinook.db")
    connection.execute("delete from customers where customerid=63")
    
    connection.commit()
    
deleteCustomer()
```

```python
#"delete" fonksiyonunun kullanimi
# db. gidip baktigimizda var olan "istanbul" isimli sehirlerin silindigini goruruz
def deleteCustomer():
    connection = sqlite3.connect("chinook.db")
    connection.execute("delete from customers where city='istanbul'")
    
    connection.commit()
    
deleteCustomer()
```

```python
#"join" fonksiyonunun kullanimi
def joinOperasyonlari():
    connection = sqlite3.connect("chinook.db")
    cursor= connection.execute("""select albums.Title,
                            artists.Name from artists 
                            inner join albums
                            on artists.ArtistId=albums.ArtistId""")
    for row in cursor:
        print("Title = "+row[0]+" Name = "+row[1])

joinOperasyonlari()
```
