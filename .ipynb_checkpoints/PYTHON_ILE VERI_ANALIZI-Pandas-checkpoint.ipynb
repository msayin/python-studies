# GroupBy İle Çalışmak

# Drop ve Axis İle Çalışmak

# Join İşlemleri

```python
# join lar iki farkli datanin belli kolonlara kriterlere gore bir araya getirilmesidir
# Genellikle iliskisel sistemlerde bunu sıklıkla  kullaniriz
# Genellikle bayanlarda evlenince soy isim falan degisiyor bu durumda bir musterinin
# Bilgilerini guncellerken sadece kimlik bilgilerinin tutuldugu dataframe i guncellemek yeterli olacak. 
# Yani sparis listesine gidip yuzlerce kez soy ismini degistirmemize gerek yok
# Ancak bu sekilde musterinin kisisel bilgileri ve siparisleri iki farkli dataframe de tutulursa bu veriler
# Bu durumda sadece musterinin isim bilgilerini yenilememiz yetecek
# O siparisbilgilerindeki tum degisikligi yapacaktir
```

```python
# join islemlerinde musterinin "ID" si uzerinden iki tadaframe deki 
# bilgilerini birlestiririz ve tek dataframe haline getiririz
```

```python
# Bu islem icin bir sozluk olusturalim
# sozluk oldugu icin datayi bir liste formatinda tutacagiz

data1= {
        'id': [1,2,3,4],
        'ad': ["Fatih", "Ayse", "Sinan", "Munevver"],
        'soyad': ["Sayin", "Sayin", "Sayin", "Bolat"]
}

data2= {
        'id': [1,3,4,7],
        'ad': ["Semih", "Ali", "Mirac", "Hasan"],
        'soyad': ["Sayin", "Sayin", "Sayin", "Bolat"]   
}
```

```python
# Bu bilgileri yani data1 ve data  yi dataframe cevirelim
# Bu durumda pandas i import etmemiz lazim
import pandas as pd

data1Df=pd.DataFrame(data1) # bu durumda bu data lari dataframe cevirmis olduk
data2Df=pd.DataFrame(data2)
print(data1Df)
print(data2Df)
```

```python
# Burada bu iki datayi join etmemiz gerekecek
# Burumda inner left yada right join ile merge etmemiz gerekecek
# Burada pandas in "pd.merge" isimli bir fonksiyonunu kullanacagiz
# Burda join islemini "ID" gore yapacagiz
# "inner join" isleminde iki tana data da belirttigimiz kolon "ID" si ayni olanlari yan  yana getir demek
# bu durumda 1,3,4 ortak noktanin var oldugu goruyoruz
# 2 ile 7 yi bir ara getiremedi cunku 1.data da 7 yok, 2.data da da 2 yok ve  eslesme saglanamiyor

print(pd.merge(data1Df,data2Df,on='id',how='inner'))

```

```python
# left join, solda olup sag da olmayanlari da getirir.
# yani burasi onemli "olmayanlari" da getiriyor
# Goruldugu gibi "inner join" aynisini getiriryor ve 
# Goruldugu gibi 2 numarali data da sag da olmadigi icin Nan olarak geldi.
# ama solda var. Ayse Sayin
print(pd.merge(data1Df,data2Df,on='id',how='left'))
```

```python
# Righ join da lef in tam tersin
# goruldugu gibi left NaN  right ta. Hasan Polat var
print(pd.merge(data1Df,data2Df,on='id',how='right'))
```

# Concat İşlemleriyle Çalışmak

```python
# "Concat" Concatined isleminden yani bir araya getirmekten bahsediyor.
# yani data lari alt alta veya ust uste getirmek, toplamak
# burada da goruldugu gibi 2 dataframe i alt alta verdi
# yeni ad soy adlari alt alta getirdi, kendi indexlerini korudu 
# bizim ID lerimizi verdi.

print(pd.concat([data1Df,data2Df]))
```

```python
# bu durumda da yatay olarak concat etti.
# bunu parcalanmis ve iliskisel datalarda kullaniyoruz
print(pd.concat([data1Df,data2Df],axis=1))
```
