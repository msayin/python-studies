# GroupBy

```python
movies=pd.read_csv("imdb-1000.csv")
print(movies.columns)
movies.head()
```

```python
#1000 movies mean
print(movies.star_rating.mean())
```

```python
# categoties of movies 
print(movies.groupby("genre").star_rating.mean())
```

```python
#star rating by duration
print(movies.mean())
```
