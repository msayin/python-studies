# Analyzing "Richest" Top 10 Economies in the World - 2019 
# Python using Matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
# Top 10 Economies in the World - 2019
x= ["1.USA","2.CHINA","3.JAPAN","4.GERMANY","5.UK","6.INDIA","7.FRANCE", "8.BRAZIL", "9.ITALY", "10.CANADA"]
y= [21.410, 15.543, 5.362, 4.416, 3.155, 3.060, 3.022, 2.261, 2.256, 1.908] # Countries by GDP (IMF) (trillion)
z= [65.064, 10.841, 42.269, 52.885, 2.302, 46.984, 44.759, 37.349, 10.693, 51.015] # GDP Per Capital (thousand)
```

```python
# Countries by GDP (IMF) (trillion)
f=plt.figure(figsize=(10,6))
plt.plot(x,y)
```

```python
# GDP Per Capital (thousand)
f=plt.figure(figsize=(10,6))
plt.plot(x,z, "r")
```

```python
# Setting the color
f=plt.figure(figsize=(10,6))
plt.plot(x,z,"y")
```

```python
# Setting the lable
f=plt.figure(figsize=(10,6))
plt.plot(x,y,"g")
plt.xlabel("Countries")
plt.ylabel("GDP")
plt.title("Countries GDP")
```

```python
# Multiple Plots
f=plt.figure(figsize=(14,5))
plt.subplot(1,2,1)
plt.plot(x,y,"r--") # settting - - - 
plt.subplot(1,2,2)
plt.plot(x,z,"y*-") # adding * -
plt.tight_layout() # setting the distance between multiple Plots 
```

```python
# Working with Objects
f=plt.figure(figsize=(10,5))
axes=f.add_axes([0.1,0.1,0.9,0.9]) # arrange the graph size
axes.plot(x,y ,"r")
axes.set_xlabel("Countries")
axes.set_ylabel("GDP")
axes.set_title("Countries GDP")
```

```python
# A Plot inside of Another Plot
f=plt.figure(figsize=(13,6))
axes1=f.add_axes([0.1,0.1,0.9,0.9]) # arrange the graph size
axes2=f.add_axes([0.5,0.5,0.4,0.4])
axes1.plot(x,y ,"r")
axes2.plot(x,z ,"y")
plt.tight_layout()
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2) # Adding to columns end rows
# first graph
axes[0].plot(x,y,"b")
axes[0].set_title("GDP")

# second graph
axes[1].plot(x,z,"y")
axes[1].set_title("GDP Per Capital")
plt.tight_layout()# Setting the distance between multiple Plots 
```

```python
# multiple functions setting

f=plt.figure(figsize=(10,5))
axes=f.add_axes([0.1,0.1,0.9,0.9]) 
axes.plot(x,y,color="red", 
    linestyle="-.", 
    linewidth=5, 
    marker="o", 
    markersize=20, 
    markerfacecolor="yellow", 
    markeredgewidth=4,
    markeredgecolor="blue", 
    alpha=0.6) 
axes.set_xlabel("Countries")
axes.set_ylabel("GDP")
axes.set_title("Countries GDP")
```

```python
f=plt.figure ()
axes=f.add_axes([0.1,0.1,0.9,0.9])
axes.plot(x,y,color="red")
axes.set_xlabel("Countries")
axes.set_ylabel("GDP")
axes.set_title("Countries GDP")
axes.set_xlim([2,6]) # setting the counties 
axes.set_ylim([2000,7500])
```

```python
# Setting the Plot size
f=plt.figure(figsize=(10,5))
axes=f.add_axes([0,0,1,1]) 
axes.plot(x,y)
axes.set_xlabel("Countries")
axes.set_ylabel("GDP")
axes.set_title("Countries GDP")
```

```python
# Setting the Multiple Plots distance
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(13,5))
# first graph
axes[0].plot(x,y,"b")
axes[0].set_title("GDP")

# seconde graph
axes[1].plot(x,z,"y")
axes[1].set_title("GDP Per Capital")
plt.tight_layout()#Setting the distance between multiple Plots       
```

```python
# Setting the Multiple lines chart and legebds
fig=plt.figure(figsize=(10,5))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,"r",label="Countries GDP")
ax.plot(x,z,"y",label="GDP Per Capital")
ax.legend(loc=0)
```

```python
fig=plt.figure(figsize=(10,5))
plt.scatter (x,y)
```

```python
# setting steps
fig=plt.figure(figsize=(10,5))
plt.step(x,y, "r")
```
