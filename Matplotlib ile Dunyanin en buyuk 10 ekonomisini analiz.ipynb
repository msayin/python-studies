```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
# Note: This list is based on World Population Review - 2019
# Gross domestic product per capita, current prices, U.S. dollars
```

```python
# "Richest" Top 10 Economies in the World - 2019
# Countries by GDP (IMF) (trillion)
# GDP Per Capital (thousand)
# Source: http://worldpopulationreview.com/countries/countries-by-gd
```

```python
# Top 10 Economies in the World - 2019
x= ["1.USA","2.CHINA","3.JAPAN","4.GERMANY","5.UK","6.INDIA","7.FRANCE", "8.BRAZIL", "9.ITALY", "10.CANADA"]
y= [21.410, 15.543, 5.362, 4.416, 3.155, 3.060, 3.022, 2.261, 2.256, 1.908] # Countries by GDP (IMF) (trillion)
z= [65.064, 10.841, 42.269, 52.885, 2.302, 46.984, 44.759, 37.349, 10.693, 51.015] # GDP Per Capital (thousand)
```

```python
# ulkelerin GSMH gore olusturuldu
fig=plt.figure(figsize=(10,5))

plt.plot(x,y)
```

```python
# Burada kisi basi gelire gore olusturuldu
fig=plt.figure(figsize=(10,5))
plt.plot(x,z, "y")
```

```python
# Rekler atamak, istediginiz rengin kodunu yaziyorsunuz
fig=plt.figure(figsize=(10,5))
plt.plot(x,z,"r")
```

```python
# Burada x,y label ve title atamis olduk
fig=plt.figure(figsize=(10,5))
plt.plot(x,y,"g")
plt.xlabel("Countries")
plt.ylabel("GDP")
plt.title("Countries GDP")
```

```python
f=plt.figure ()
axes=f.add_axes([0.1,0.1,0.9,0.9]) # Burada frafigin cercevesi icin boyutlarini vererek bir tablo olusturduk
axes.plot(x,y,color="red", 
    linestyle="-.", # burada cizgi olsun dedik
    linewidth=5, # burada cizginin kalinligini belirledik
    marker="o", # burada daireler ekledik
    markersize=20, # burada dairelere boyutunu belirledik
    markerfacecolor="yellow", #burada dairelere rengini belirledik
    markeredgewidth=4, # burada dairelere kenar kalinligini belirledik
    markeredgecolor="blue", # burada dairelere kenar rengini belirledik
    alpha=0.6) # burada grafigin saydamligini belirledik
axes.set_xlabel("Countries")
axes.set_ylabel("GDP")
axes.set_title("Countries GDP")
```

```python
f=plt.figure ()
axes=f.add_axes([0.1,0.1,0.9,0.9]) # Burada frafigin cercevesi icin boyutlarini vererek bir tablo olusturduk
axes.plot(x,y,color="red")
axes.set_xlabel("Countries")
axes.set_ylabel("GDP")
axes.set_title("Countries GDP")
axes.set_xlim([2,5]) # burada kactane ulkeyi alacaksan onu belirtiyorsun, yani zumlama yapiyorsun
axes.set_ylim([2000,7500])
```
